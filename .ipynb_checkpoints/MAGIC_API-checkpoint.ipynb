```python
# -*- coding: utf-8 -*-
"""
Created on Thu Feb 28 17:47:04 2019

@author: Disus
"""

import requests
import numpy as np
import pandas as pd
import time as time
import math
from time import sleep
import datetime

start_time = time.time()


name_list = []
text_list = []
type_list = []
cmc_list = []
mana_list = []
r_list = []
g_list = []
w_list = []
b_list = []
u_list = []
non_list = []
color_list = []

result = None
x = 0
while result is None:
    try:
        response = requests.get('https://api.magicthegathering.io/v1/cards?PageSize=500&page=1')
        tcount = (response.headers['Total-Count'])
        psize = (response.headers['Page-Size'])
        tpages = math.ceil(int(tcount)/int(psize))
        print("Total de {0} cards".format(tcount))
        contador = 5
        result = 1
    except:
        print("Tentativa n:{0}".format(x))
        sleep(2)
        x += 1
        pass
for j in range(tpages+1):
    result = None
    sleep_time = 2
    while result is None:
        try:
            url = 'https://api.magicthegathering.io/v1/cards?PageSize=100&page=' + str(j)

            response = requests.get(url)
            ratelimit = (response.headers['Ratelimit-Remaining'])
            data = response.json()
            cards = data['cards']  
            
            for i in cards:
                name_list.append(i['name'])
                type_list.append(i['types'][0])
                cmc_list.append(i['cmc'])
                try:
                    mana_list.append(i['manaCost'])
                except:
                    mana_list.append('')
                colors = []
                cores = ""
                r_list.append(0)
                g_list.append(0)
                w_list.append(0)
                b_list.append(0)
                u_list.append(0)
                non_list.append(1)
                for c in i['colorIdentity']:
                    if c == 'R':
                        r_list.pop()
                        r_list.append(1)
                        non_list.pop()
                        non_list.append(0)
                    elif c == 'G':
                        g_list.pop()
                        g_list.append(1)
                        non_list.pop()
                        non_list.append(0)
                    elif c == 'W':
                        w_list.pop()
                        w_list.append(1)
                        non_list.pop()
                        non_list.append(0)
                    elif c == 'B':
                        b_list.pop()
                        b_list.append(1)
                        non_list.pop()
                        non_list.append(0)
                    elif c == 'U':
                        u_list.pop()
                        u_list.append(1)
                        non_list.pop()
                        non_list.append(0)
                    else:
                        non_list.pop()
                        non_list.append(1) 
                    cores += c+' '
                color_list.append(cores.strip())
                try:
                    text_list.append(i['text'])
                except:
                    text_list.append('')

            perc = ((j/len(range(1,tpages))))*100
            ETA = str(datetime.timedelta(seconds=(time.time() - start_time)/(j+1)*(tpages-j)))
            print('{0} de {1} - Perc:{2:8.2f}% - ETA: {3}'.format(j,tpages,perc, ETA), end='\r')
            result = 1
            x = 0
            sleep_time = 2

        except:
            print("Tentativa n:{0}. Esperando {1}seg para tentar novamente (Pagina {2})".format(x+1,sleep_time,j))
            sleep(sleep_time)
            sleep_time += 1
            x += 1
            pass


print(str(datetime.timedelta(seconds=(time.time() - start_time))))

lista = list(zip(name_list,type_list,text_list,cmc_list,mana_list,color_list,r_list,g_list,w_list,b_list,u_list,non_list) )

np_lista = np.array(lista)
colunas= ['Name','Type','Text','CMC','Mana','Color','R','G','W','B','U','Incolor']
lista_pd = pd.DataFrame(data=np_lista[0:,0:],columns=colunas[0:])
lista_clean = lista_pd.drop_duplicates(inplace=False)
display(lista_clean.head())
lista_clean.to_csv('MTG_CARDS.csv', sep=';', encoding = "utf-8")
```
